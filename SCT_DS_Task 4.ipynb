```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('AccidentsBig.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['Time']=pd.to_datetime(df['Time'],errors='coerce')
df['Hour']=df['Time'].dt.hour
df[['Time','Hour']].head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Hour',data=df)
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='Weather_Conditions',data=df, order=df['Weather_Conditions'].value_counts().index)
plt.title('Accidents by Weather Conditions')
plt.xlabel('Count')
plt.ylabel('Weather Conditions')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='Road_Surface_Conditions',data=df, order=df['Road_Surface_Conditions'].value_counts().index)
plt.title('Accidents by Road Conditions')
plt.xlabel('Count')
plt.ylabel('Road Conditions')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Accident_Severity',data=df)
plt.title('Accidents Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['longitude'],df['latitude'],alpha=0.4)
plt.title('Accidents Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
numeric_df=df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Between Numeric Features')
plt.show()
```

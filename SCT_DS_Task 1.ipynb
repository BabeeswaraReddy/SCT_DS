```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
age_groups = ['0 to 20 Years', '21 to 64 Years', '65+ Years']
population_millions = [512, 807, 98]
percentages = [36.1, 57.0, 6.9]
colors = ['gold', 'skyblue', 'lightcoral']
total_population = 1420
```

```python
plt.figure(figsize=(8,5))
bars = plt.bar(age_groups, population_millions, color=colors)
```

```python
for bar, pct in zip(bars, percentages):
  height = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2, height + 10, f'{pct}%', ha='center', fontsize=12)

plt.title("India's Population Distribution by Age in 2022")
plt.xlabel("Age Group")
plt.ylabel("Population (in Millions)")
plt.ylim(0, max(population_millions) + 100)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
np.random.seed(42)
samples_0_20 = np.random.randint(0, 21, size=5120)
sample_21_64 = np.random.randint(21, 65, size=8070)
samples_65_plus = np.random.randint(65, 101, size=980)
```

```python
age_samples = np.concatenate([samples_0_20, sample_21_64, samples_65_plus])
```

```python
plt.figure(figsize=(10,6))
plt.hist(age_samples, bins=20, color='mediumseagreen', edgecolor='black')
plt.title("Simulated Age Distribution Histogram for India (2022)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
